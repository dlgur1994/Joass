```
#feature추출/ wav -> npz
import librosa
import numpy as np

#feature 추출하는 module
def extract_feature(file_name):
    X, sr = librosa.load(file_name)
    mfccs = librosa.feature.mfcc(y=X, sr=sr, n_mfcc=39, hop_length=int(sr*0.01), n_fft=int(sr*0.02)).T
    return mfccs

n = 39
dirname = '/gdrive/My Drive/조아쓰/자료조사/18.11.07_녹음/'
save_path = '/gdrive/My Drive/19-1/training_set/'
classes = ['뉴스','리모컨','소리작게','소리크게','시간','오늘날씨','오늘일정','지니야','클로바']

#training_set 제작
for filename in classes:
  f = dirname + filename + '70.wav'#load file  
  all_mfccs = np.ndarray(shape = [0,n], dtype = np.float32)
  feature = extract_feature(f)

  all_mfccs = np.append(all_mfccs, feature, axis = 0)
  np.savez(save_path + filename + '.npz', X = all_mfccs)
  
  print(filename)
```

```
#각 npz 파일을 합쳐서 하나의 set으로
#_*_coding: utf-8_*_

import numpy as np

tc = []
td = []

N = 39
ans = np.zeros(9)
classes = ['뉴스','리모컨','소리작게','소리크게','시간','오늘날씨','오늘일정','지니야','클로바']

saved_path = '/gdrive/My Drive/19-1/training_set/'

for i in range(9):
  filename = classes[i]
  f = np.load(dirname + filename + ".npz")#load file
  
  #length of X
  num = len(f['X'])
  
  #label
  ans = np.zeros(9)
  ans[i] = 1
    
  #td = label
  for m in range(num):
    if i == 0:
      if m == 0:
        td = np.array(ans)
      else:
        td = np.vstack([td, ans])
    else:
      td = np.vstack([td, ans])

  #tc = feature
  if i == 0:
    tc = f['X']
  else:
    tc = np.vstack([tc, f['X']])

  tc = tc.astype(np.float32)
  td = td.astype(np.float32)
  
#저장
np.savez(saved_path + 'training_set.npz', X = tc, Y = td)
  
print("finish")
```
